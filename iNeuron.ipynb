One Hot Encoding

```python
#Ecoding the categorical variables.
#Standardizing the features of a dataset so that they have a mean of zero and a standard deviation of one.
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('ct',OneHotEncoder(sparse=False,drop='first'),['sex','smoker','region'])],remainder='passthrough')
```

```python
#Splitting the dataset into training set and testing set.
X=dataset.drop(['expenses'],axis=1)
Y=dataset['expenses']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size = 0.2,random_state=0)
```

Multiple Linear Regressor

```python
from sklearn.linear_model import LinearRegression
#fitting multiple linear regression on this data.
```

```python
from sklearn.pipeline import Pipeline
pipe1=Pipeline(steps=[
    ("step1",ct),
    ("step2",StandardScaler()),
    ("step3",LinearRegression()),
])
```

```python
pipe1.fit(X_train,Y_train)
```

```python
#predicting y test values after fitting the multiple linear regression.
y_pred1=pipe1.predict(X_test)
print(y_pred1)
```

```python
#Root mean sq error for Multiple linear regression
import math
RMSE_test1=math.sqrt(sum((Y_test-y_pred1)**2)/(len(y_pred1)))
print(RMSE_test1)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#Adjusted r sq value.
R2Score_model_test1 = r2_score(Y_test, y_pred1)
print(R2Score_model_test1)
```

Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
#fitting random forest regression on this data.
```

```python
pipe2=Pipeline(steps=[
    ("step1",ct),
    ("step2",StandardScaler()),
    ("step3",RandomForestRegressor(n_estimators = 250, random_state = 221)),
])
```

```python
# fit the regressor with x and y data
pipe2.fit(X_train, Y_train)
```

```python
y_pred2=pipe2.predict(X_test)
print(y_pred2)
```

```python
#Root mean sq error Random forest regression
import math
RMSE_test2=math.sqrt(sum((Y_test-y_pred2)**2)/(len(y_pred2)))
print(RMSE_test2)
```

```python
#Adjusted R sq 
R2Score_model_test2 = r2_score(Y_test, y_pred2)
print(R2Score_model_test2)
```

Gradiant Boosting


```python
from sklearn.ensemble import GradientBoostingRegressor
#Fitting the same data using gradiant boosting.
```

```python
pipe3=Pipeline(steps=[
    ("step1",ct),
    ("step2",StandardScaler()),
    ("step3",GradientBoostingRegressor(learning_rate=0.01,
                                n_estimators=330,
                                random_state = 221))
])
```

```python
pipe3.fit(X_train,Y_train)
```

```python
y_pred3=pipe3.predict(X_test)
print(y_pred3)
```

```python
#RMSE
RMSE_test3=math.sqrt(sum((Y_test-y_pred3)**2)/(len(y_pred3)))
print(RMSE_test3)
```

```python
#Adjusted R sq.
R2Score_model_test3 = r2_score(Y_test, y_pred3)
print(R2Score_model_test3)
```

We can observe that for Gradient boosting regression helps us to achieve the highest possible adjusted R sq value. Whereas, Random forest regression is also giving a higher adjusted R sq. value not as good as compared to the Gradient boosting regression. However, Multiple linear regression has a very less adjusted R sq value so for prediction purpose it will fail to predict accurately.
